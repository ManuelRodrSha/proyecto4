# links (se va a convertir en  un dataframe con más columnas y menos rows. Para ello se va a hacer una operación de pivot)

```python
df_links = pd.read_csv("..\\datasets\\links.csv")
df_types = pd.read_csv("..\\datasets\\link_types.csv")
```

```python
# Paso 1: Merge para tener los nombres de los tipos
df_merged = pd.merge(df_links, df_types, on='link_type_id', how='left')

# Paso 2: Pivot para convertir tipos en columnas
df_pivot = df_merged.pivot(index='show_id', columns='link_type', values='link')

# Paso 3: Renombrar columnas a "link_tipo"
df_pivot.columns = [f'link_{col}' for col in df_pivot.columns]

# Paso 4: Reset index si querés que 'show_id' sea columna
df_final = df_pivot.reset_index()

df_final
```

```python
columnas_extra = ['link_backdrop_path', 'link_homepage', 'link_poster_path']
shows_df = pd.merge(shows_df, df_final[['show_id'] + columnas_extra], on='show_id', how='left')
shows_df.head(5)
```

```python
##guardamos el dataset
shows_df.to_csv("..\\datasets\\shows_filtrado.csv", index=False) 
```
